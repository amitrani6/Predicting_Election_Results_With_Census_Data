```python
import requests
import pandas as pd
import ast
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
```

```python
congressional_district_url = 'https://api.census.gov/data/2015/acs/acs1?get=B00001_001E,NAME&for=congressional%20district:*&in=state:*'
congressional_district_content = requests.get(congressional_district_url)
congressional_district_list = ast.literal_eval(congressional_district_content.text)
print(congressional_district_list)
```

```python
#This block of code removes the unecessary code
congressional_df = pd.DataFrame(congressional_district_list[1:], columns=congressional_district_list[0])
congressional_df
```

```python
congressional_df['State'] = congressional_df.NAME.apply(lambda x: x.split(', ')[1])
```
